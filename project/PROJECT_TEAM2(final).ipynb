```python
feat_labels=['Age','Gender','Total_Bilirubin','Alkaline_Phosphotase','Aspartate_Aminotransferase','Total_Protiens','Albumin_and_Globulin_Ratio']
```

```python
from sklearn.feature_selection import SelectFromModel
```

```python
dataset1=dataset[["Age","Gender","Total_Bilirubin","Alkaline_Phosphotase","Aspartate_Aminotransferase", "Total_Protiens", "Albumin_and_Globulin_Ratio","Dataset"]]
dataset1
```

```python
x=dataset1.iloc[:,:7].values
x
```

```python
y=dataset1.iloc[:,7:].values
y
```

```python
import sklearn
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb=LabelEncoder()
x[:,1]=lb.fit_transform(x[:,1])
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.07,random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=10,random_state=40,n_jobs=-1)
```

```python
rf.fit(x_train,y_train)
```

```python
for feature in zip(feat_labels,rf.feature_importances_):
    print(feature)
```

```python
y_pred=rf.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
